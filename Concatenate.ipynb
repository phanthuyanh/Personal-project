```python
import pandas as pd
```

```python
p15 = pd.read_csv('project_15.csv')
p8 = pd.read_csv('project_8.csv')
p13 = pd.read_csv('project_13.csv')
```

```python
combined = pd.concat([p15, p8, p13], ignore_index=True)
combined['Transaction Date'] = pd.to_datetime(combined['Transaction Date'])
combined['Month'] = combined['Transaction Date'].dt.month
```
